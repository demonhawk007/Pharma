```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
% matplotlib inline
np.set_printoptions(formatter={'float_kind':lambda x: "%.2f" % x})
```

```python
train_df= pd.read_csv('counterfeit_train.csv')
train_df['label'] = 'train'
test_df=pd.read_csv("counterfeit_test.csv")
test_df['label'] = 'test'
```

```python
train_df=train_df.drop('Medicine_ID',axis=1)
```

```python
test_df=test_df.drop('Medicine_ID',axis=1)
```

```python
#Concat the datasets
concat_df = pd.concat([train_df , test_df])
```

```python
features_df = pd.get_dummies(concat_df,columns=['DistArea_ID',"Active_Since",'Medicine_Type','SidEffect_Level','Area_Type',
                                               'Area_City_Type','Area_dist_level'])
```

```python
type(features_df)
```

```python
train_df
```

```python
# Split your data
train_df = features_df[features_df['label'] == 'train']
test_df = features_df[features_df['label'] == 'test']
```

```python
# Drop your labels
train_df = train_df.drop('label', axis=1)
test_df = test_df.drop('label', axis=1)
```

```python
train_df.head()
```

```python
train_df['Counterfeit_Sales1']=train_df['Counterfeit_Sales']
train_df=train_df.drop(['Counterfeit_Sales'],axis=1)
```

```python
x=train_df.iloc[:,0:-1].values
y=train_df.iloc[:,-1].values
```

```python
#data preprocessing
```

```python
from sklearn.preprocessing import Imputer
```

```python
imputer=Imputer()
```

```python
imputer=imputer.fit(x[:,[0,1,2]])
```

```python
x[:,[0,1,2]]=imputer.transform(x[:,[0,1,2]])
```

```python
print(x)
```

```python
#splitting the dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
regressor=RandomForestRegressor(n_estimators=10,criterion='mae',random_state=1)
```

```python
regressor.fit(x,y)
```

```python
#Predicting on test
```

```python
y_pred=regressor.predict(x_test)
```
